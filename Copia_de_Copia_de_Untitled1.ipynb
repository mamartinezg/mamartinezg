
En el siguiente código se filtra los registros que tienen Fuente_Datos igual a 'MATRICULADO', y luego realiza el proceso de eliminación de duplicados para mantener el último Año_Periodo en esos registros.

```python
df = df[df['Fuente_Datos'] == 'MATRICULADO']
df['Año_Periodo'] = pd.to_datetime(df['Año_Periodo'])
df.sort_values(by=['Id_Usuario', 'Año_Periodo'], ascending=[True, False], inplace=True)
df.drop_duplicates(subset='Id_Usuario', keep='first', inplace=True)
df.reset_index(drop=True, inplace=True)
```

Ahora se procede a realizar el cruce de las dos bases de datos para esto se toma como llave programa academico y año, pues estos registros estan en ambas bases de datos.

```python
# Realiza una combinación basada en las columnas llave es decir programa academico y año
VA['Estu Fechanacimiento'] = pd.to_datetime(VA['Estu Fechanacimiento'])
df['Fecha_Nacimiento'] = pd.to_datetime(df['Fecha_Nacimiento'],errors='coerce')
df = df.drop_duplicates(subset=['Fecha_Nacimiento', 'Nombre_Programa'])
DataVA = VA.merge(df, left_on=['Estu Fechanacimiento', 'Estu Prgm Academico'],
                  right_on=['Fecha_Nacimiento', 'Nombre_Programa'], how='left')
```

```python
DataVA_Final['Promedio_Acumulado'] = pd.to_numeric(DataVA_Final['Promedio_Acumulado'], errors='coerce')
```

```python
# Primero, calcula los promedios por la combinación única de 'Anosbpro' y 'Estu Prgm Academico' solo para las filas con valores no perdidos en 'Promedio_Acumulado'
promedios = DataVA_Final.dropna(subset=['Promedio_Acumulado']).groupby(['Anosbpro', 'Estu Prgm Academico'])['Promedio_Acumulado'].mean().reset_index()
promedios.head()
```

```python
# Luego, crea un diccionario con estos valores promedio
promedio_imputacion = dict(zip(zip(promedios['Anosbpro'], promedios['Estu Prgm Academico']), promedios['Promedio_Acumulado']))
promedio_imputacion
```

```python
# Ahora, utiliza el método fillna para imputar los valores perdidos en 'Promedio_Acumulado' con los promedios correspondientes
DataVA_Final['Promedio_Acumulado'].fillna(DataVA_Final.apply(lambda row: promedio_imputacion.get((row['Anosbpro'], row['Estu Prgm Academico']), np.nan), axis=1), inplace=True)
```
